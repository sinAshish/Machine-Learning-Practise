```python
def each_percent(col_name):
    return train[col_name].value_counts()/len(train[col_name])
```

```python
def plot_bivariate(col_name):
    plt.figure(figsize=(12,6))
    sns.countplot(x=col_name,data=train,hue='Criminal')
    plt.show()
```
